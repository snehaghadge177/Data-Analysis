```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart.csv")
df.head(5)
```

```python
df.shape
```

## Distribution plot

```python
# Distribution plot for 'Age'

sns.distplot(df['Age'], kde  = True, hist = True)
```

```python
# distribution plot for 'RestingBP'


sns.distplot(df['RestingBP'], kde  = True, hist = True)
```

```python
# distribution plot for 'Cholesterol'

sns.distplot(df['Cholesterol'], kde  = True, hist = True)
```

```python
# Distribution plot for 'MaxHR'

sns.distplot(df['MaxHR'], kde  = True, hist = True)
```

## Pie chart

```python
# Pie Chart for 'Sex'

df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.1F')
```

```python
# pie chart for 'ChestPainType'


df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1F')

# Analysis - Most people have ASY (Asymptomatic) chest pain
```

```python
# Pie chart for 'RestingECG'

df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.1F')

# Analysis - Most people's resting electrocardiogram results are Normal
```

```python
# Pie chart for 'ST_Slope'

df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1F')

# Analysis - Most of the people have the slope of the peak exercise ST segment as 'FLAT'  
```

```python
# pie chart for 'HeartDisease'


df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1F')
```

## Violin plot

```python
sns.violinplot(df['Age'])
```

```python
# Violin plot with Sex and HeartDisease

sns.violinplot(y = df['Sex'], x = df['HeartDisease'])

# Analysis - More number of Men are having HeartDisease and
# Less number of women are having HeartDisease
```

```python
# Violin plot with Age and HeartDisease

sns.violinplot(y = df['Age'], x = df['HeartDisease'])

# Analysis - Average age of people having heart disease is 60 
```

```python
# Violin plot with Age and HeartDisease

sns.violinplot(x = df['ChestPainType'], y = df['HeartDisease'])

# Analysis - Out of the people positi8ve for a heart disease, more people are having ASY chest pain type
# And very less people having ATA chest pain type.
```
