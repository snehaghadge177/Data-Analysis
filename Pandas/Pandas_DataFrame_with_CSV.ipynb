### Fillna


```python
link = 'https://raw.githubusercontent.com/AshishJangra27/Data-Analysis-with-Python-GFG/main/3.%20Data%20Preprocessing%20-%20Removing%20Null%20Value%20Rows/googleplaystore.csv'
df = pd.read_csv(link)
df.head()
```

```python
df.isnull().sum()
```

```python
mean = round(df['Rating'].mean(), 2)
mean
```

```python
df['Rating'] = df['Rating'].fillna(mean)
```

```python
df.isnull().sum()
```

### Remove the index

```python
dct = {'name': ['A','B','C'],
      'age' : [21,23,26]}
df1 = pd.DataFrame(dct)
df1
```
