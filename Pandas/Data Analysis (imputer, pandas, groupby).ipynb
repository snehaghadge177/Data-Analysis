```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Create an imputer function
impute = Imputer(missing_values = np.nan ,  strategy = 'mean')
```

 Extract a df to apply imputer on.

df.iloc[ : , 2:3].values

```python
# Fit the imputer function to extracted iloc
impute.fit(df.iloc[ : , 2:3].values)
```

```python
# Use Transform function to finally make changes
df.iloc[ : , 2:3] = impute.transform(df.iloc[ : , 2:3].values)
```

```python
# Now check that all the null values from Rating column are gone!
df.isnull().sum()
```

```python
# Drop other rows as they are less in number and won't impact our dataset

df = df.dropna()
```

### Q1. How many FREE apps are there in ART_AND_DESIGN?

```python
# Using our own logic

count = 0
for i in df:
    if(i[1] == 'ART_AND_DESIGN' and i[6] == 'Free'):
        count += 1
        print(i[0], " : ", i[1]," : ", i[6])
print(count)
```

```python
# Using pandas

df_pr = df[df['Type'] == 'Free']
df_pr[df_pr['Category'] == 'ART_AND_DESIGN']
```
